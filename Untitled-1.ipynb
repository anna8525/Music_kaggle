```python
X = music_train.drop('music_genre', axis=1)
y = music_train['music_genre']

num_features = X.select_dtypes(include='number').columns.tolist()

categ_features = X.select_dtypes(exclude='number').columns.tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
pip install feature-engine
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
num_pipeline = Pipeline(steps=[
    ('impute', IterativeImputer(max_iter=10, random_state=42)),
    ('scale', RobustScaler())
    ])
cat_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('encode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
col_trans = ColumnTransformer(transformers=[
    ('num_pipeline',num_pipeline,num_features),
    ('cat_pipeline',cat_pipeline,categ_features)
    ],
    remainder='passthrough',
    n_jobs=-1)
```

LogisticRegression - Логистическая регрессия. Начнем именно с нее.

```python
log = LogisticRegression(tol=0.01)

pipeline = Pipeline(steps=[
     ('col_trans', col_trans),
     ('model', log)
 ])

param_dist = {
     'model__C': np.logspace(-1, 1, 2),
     'model__penalty': ['l1', 'l2','elasticnet'],
     'model__solver': ['liblinear', 'saga', 'sag']
 }

grid = RandomizedSearchCV(pipeline,
                     param_dist,
                     n_iter=30,
                     cv=5,
                     verbose=5,
                     random_state= 123,
                     scoring='f1_micro',
                     n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
predicted = grid.predict(X_test)
f1_score(y_test, predicted, average='micro').round(2)
```

DecisionTreeClassifier - деревo решений.


```python
DTC = DecisionTreeClassifier()

pipeline = Pipeline(steps=[
     ('col_trans', col_trans),
     ('model', DTC)
 ])

param_dist = {
    'model__criterion':  ["gini", "entropy"],
    'model__max_depth': [2,4,10],
    'model__min_samples_leaf' : [30, 60, 10],
    'model__min_samples_split':[1,2,3]
 }

grid = RandomizedSearchCV(pipeline,
                     param_dist,
                     n_iter=30,
                     cv=5,
                     verbose=5,
                     random_state= 123,
                     scoring='f1_micro',
                     n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
predicted = grid.predict(X_test)
f1_score(y_test, predicted, average='micro').round(2)
```
